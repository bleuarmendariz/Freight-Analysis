# Pull Data and make some basic plots

Cass Freight Index: Shipments -  is a record of volume. Its important to note that it includes all forms of transportation - TL, LTL, Flatbed, Dryvan etc. This matters becuase different modes of transportation tend to have different types of seasonality. 

Cass Freight Index: Expenditures - is a record how how much is spent on freight. 

I want to make a basic plot of all the data we have from FRED for specifically shipments for now - just purely to make sure everything is working, and brainstorm about how we can depict seasonalities of the market

```python
cass_shipments = pd.DataFrame(fred.get_series(series_id = 'FRGSHPUSM649NCIS'))
```

```python
cass_shipments.plot(figsize = (10,5), title = 'Cass Freight Index', lw = 2)
plt.show()
```

```python
cass_shipments
```

After throwing the data into a quick plot, and displaying the dataframe, I have a few ideas. 

1. Break data apart by year, so we can compare each year
2. Get the highest volume months, and lowest volume months per year

# Grouping Data by Year

Doing some basic cleaning here to create a date column, rename some columns, and get datatyps right.

```python
cass_shipments['Date'] = cass_shipments.index
cass_shipments['Date'] = pd.to_datetime(cass_shipments['Date'])
cass_shipments = cass_shipments.reset_index(drop = True)
```

```python
cass_shipments
```

```python
cass_shipments = cass_shipments.rename(columns = {0:"Values"})
```

```python
cass_shipments
```

I want to seperate the date column into Year and Month, and then map the number we get for month to an actual month name. 

```python
cass_shipments['Year'] = cass_shipments['Date'].dt.year
cass_shipments['Month'] = cass_shipments['Date'].dt.month
```

```python
month_conversion = {1:"Jan", 2:"Feb", 3:"Mar", 4:"Apr", 5:"May", 6:"Jun", 7:"Jul", 8:"Aug", 9:"Sep", 10:"Oct", 11:"Nov", 12:"Dec"}
```

```python
cass_shipments['Month'] = cass_shipments['Month'].map(month_conversion)
```

```python
cass_shipments
```

```python
cass_2022 = cass_shipments[cass_shipments['Year'] == 2022]
```

```python
cass_2022.plot('Month', 'Values')
plt.show()
```

Now that the dataframe is cleaned up a bit, I want to start on my first idea which is to compare years to see market trends. I personally like gridding out data like this, but for now I think stacking this data would be a quicker way to actually see the seasonality insight

```python
years = [2016,2017,2018,2019,2020,2021,2022]
```

```python
plt.figure(figsize = (10,5))
for i in years:
    cass_plot = cass_shipments[cass_shipments['Year'] == i]
    plt.plot(cass_plot['Month'], cass_plot['Values'], label = i)

plt.legend(prop = {'size':6})
plt.title('CASS Freight Shipment Trends')
plt.show()
```

This graph is perfect. We can see Year vs Year for the past few years how freight shipments are trending, and even how each year is performing vs other years in freight

By looking at this data, we can see a few insights:

> Freight shipments typically peak in May. After the big peak in may the usually will come down during the rest of the summer, and peak again in August.

> After the August peak they typically decline to the end of the year. This is suprising due to how many holidays there are at the end of the year. The shipments in August are most likey for the EOY rush - Black Friday, Thanksgiving, Christmas, etc

> Freight shipments are usually at their lowest at the start of the year - following the decline from the previous year

I find it hard, personally, to see what years were doing better than other years overall. I want to get the average rating per year, and create one plot with that. 

```python
cass_plot_means = {}
for i in years:
    cass_plot = cass_shipments[cass_shipments['Year'] == i]
    cass_plot_means[i] = float(cass_plot['Values'].mean())
```

```python
cass_plot_means = pd.DataFrame.from_dict(cass_plot_means, orient = 'index')
cass_plot_means = cass_plot_means.reset_index()
```

```python
cass_plot_means
```

```python
cass_plot_means = cass_plot_means.rename(columns = {'index':'Year', 0:"Mean"})
```

```python
cass_plot_means
```

```python
cass_plot_means.plot('Year', 'Mean')
plt.title("Mean Volumes Per Year")
plt.show()
```

2018 is one of the best freight years in our data in terms of volume of shipments. The story this tells us overall though, is that volume had been steadily increasing with a high in 2018. It started to decline in 2019, and then crash at 2020 - most likley due to COVID conditions. Since then the market has bounced back, but has still yet to match 2018 highs.
